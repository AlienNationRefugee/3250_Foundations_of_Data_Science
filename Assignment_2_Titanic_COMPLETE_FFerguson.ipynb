# Let's try building a Gradient Boosting and Random Forest in H2O

```python
import h2o
h2o.init(max_mem_size="2G")
h2o.remove_all()

from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.estimators.random_forest import H2ORandomForestEstimator
from h2o.model.metrics_base import H2OMultinomialModelMetrics
from h2o.estimators.glm import G

```

We import the reduced Titanic dataset (891k rows, 9 columns, 4 numerical, 5 categorical) and then split the data 3 ways:  
    70% for training  
    20% for validation (hyper parameter tuning)  
    10% for final testing  

We will train a data set on one set and use the others to test the validity of the model by ensuring that it can predict accurately on data the model has not been shown.  The second set will be used for validation most of the time.  The third set will be withheld until the end, to ensure that our validation accuracy is consistent with data we have never seen during the iterative process. 

```python
#Import the Titanic dataframe into an H2O dataframe
hf = h2o.H2OFrame(df)

#Reorder columns with response variable at the end dropping PassengerId
hf=hf[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Survived']]
hf
```

```python
#split out training, validation and testing frames (70/20/10)
trainf, validf, testf=hf.split_frame([.7,.2],seed=314159)
```

```python
#Prepare explanatory columns
hf_X=hf.col_names[:-1]
hf_Y=hf.col_names[-1]
```

Build Random Forest model with the following parameters:
 
 **model_id:** Not required, but allows us to easily find our model in the [Flow](http://localhost:54321/) interface  
 **ntrees:** Maximum number of trees used by the random forest. Default value is 50. We can afford to increase this, as our early-stopping criterion will decide when the random forest is sufficiently accurate.  
 **stopping_rounds:** Stopping criterion described above. Stops fitting new trees when 2-tree rolling average is within 0.001 (default) of the two prior rolling averages. Can be thought of as a convergence setting.  
 **score_each_teration:** predict against training and validation for each tree. Default will skip several.  
 **seed:** set the randomization seed so we can reproduce results

```python
#Random Forest Model Parameters
rf_Titanic_v1 = H2ORandomForestEstimator(
    model_id="rf_Titanic_v1",
    ntrees=200,
    stopping_rounds=2,
    score_each_iteration=True,
    seed=1000000)
```

```python
#Random Forest Model Construction
rf_Titanic_v1.train(hf_X, hf_Y, training_frame=trainf, validation_frame=validf)
```

```python
#Take a look at the RF model overview
rf_Titanic_v1
```

The most important explanatory variables for survival are: 1) Sex, 2) Age, 3) Fare and 4) Pclass

```python
#Take a look at the RF validation statistics
rf_Titanic_v1.score_history()
```

```python
rf_Titanic_v1
```

Build Gradient Boosting model with the following parameters:
 
 **model_id:** Not required, but allows us to easily find our model in the [Flow](http://localhost:54321/) interface  
 **seed:** set the randomization seed so we can reproduce results

```python
gbm_Titanic_v1 = H2OGradientBoostingEstimator(
    model_id="gbm_Titanic_v1",
    seed=2000000
)
gbm_Titanic_v1.train(hf_X, hf_Y, training_frame=trainf, validation_frame=validf)
```

```python
#Take a look at the GBM model overview
gbm_Titanic_v1
```

```python
#Take a look at GBM validation statistics
gbm_Titanic_v1.score_history()
```
